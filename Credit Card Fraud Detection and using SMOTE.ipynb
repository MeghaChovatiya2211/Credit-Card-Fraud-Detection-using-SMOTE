### Step 5: Splitting data into Train and Test Set

```python
from sklearn.model_selection import train_test_split

df1 = df
x = df1.drop(['Class'], axis=1)
y = df1['Class']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state = 42)
```

### Step 6: Data Scaling 

```python
#Normalizing the data 

from sklearn.preprocessing import MinMaxScaler

norm = MinMaxScaler()

x_train =  norm.fit_transform(x_train)
x_test = norm.fit_transform(x_test)
```

### Step 7: Dealing with the imbalanced data using SMOTE 

```python
print("y_train: ",y_train.shape)
print("y_test: ",y_test.shape)

print("\n")

print("Count of 0s in y_train: ", sum(y_train==0))
print("Count of 1s in y_test: ", sum(y_train==1))
```

```python
#We balance the train data only by using sampling techniques 
#While applying SMOTE, we apply to entire train data including x_train and y_train
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train, y_train = smote.fit_resample(x_train, y_train.ravel())

#ravel() - changes a 2-dimensional array or a multi-dimensional array into a contiguous flattened array
```

```python
print("Total Training Rows: ", y_train.shape)
print("Count of 0s in y_train after applying SMOTE: ", sum(y_train==0))
print("Count of 1s in y_test after applying SMOTE: ", sum(y_train==1))

#Now the count of 0s and 1s is the same
```

### Step 8: Applying ML Algorithm 

```python
#Apply Machine Learning Models

from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(random_state=0)
tree.fit(x_train, y_train)

from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
forest.fit(x_train, y_train)
    

print("[1]Decision Tree :", tree.score(x_train, y_train))
print("[2]Random Forest :", forest.score(x_train, y_train))
```

```python
y_pred = forest.predict(x_test)
```

```python
#Creating Confusion Matrix 
from sklearn.metrics import confusion_matrix

sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt ='d', cmap = "Greens")
```

```python
from sklearn.metrics import classification_report

TN, FP, FN, TP = confusion_matrix(y_test, y_pred).ravel()

recall = TP / (TP + FN)
print("recall :", recall)

precision = TP / (TP + FP)
print("Precision :", precision)

f1_score = 2*((precision*recall)/(precision+recall))
print("f1 score :", f1_score)

matrix = classification_report(y_test, y_pred, labels=[1,0])
print('\n')
print('Classification Report :\n', matrix)
```
